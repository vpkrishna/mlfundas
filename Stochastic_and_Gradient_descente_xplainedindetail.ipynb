```python
import numpy as np
import math
import random

############################################################
# Optimization problem

trueW = np.array([1, 2, 3, 4, 5])
def generate():
    random.seed(10)
    x = np.random.randn(len(trueW))
    y = trueW.dot(x) + np.random.randn()
    print('example data ==> ', x, y)
    return (x, y)

trainExamples = [generate() for i in range(6)]

def phi(x):
    return np.array(x)

def initialWeightVector():
    return np.zeros(len(trueW))

##### In gradient descent
### 1) Dot product of input with weights is taken w.dot(phi(x))
### 2) 2 * (w.dot(phi(x)) - y) * phi(x) is calculated  ie   gradient calculated as first derivative of MSE
### 3) Sum of above 2 is taken, averaged and a single gradient is calculated for each of the feature

def trainLoss(w):
    return 1.0 / len(trainExamples) * sum((w.dot(phi(x)) - y)**2 for x, y in trainExamples)

def gradientTrainLoss(w):
    return 1.0 / len(trainExamples) * sum(2 * (w.dot(phi(x)) - y) * phi(x) for x, y in trainExamples)

###In stochastic gradient descent
### 1) gradient  is calculated for each record.. Rest of claculation same as above
####  ie 2 * (w.dot(phi(x)) - y) * phi(x)  [No sum and hence no x, y in trainExamples , No average]
def loss(w, i):
    x, y = trainExamples[i]
    return (w.dot(phi(x)) - y)**2

def gradientLoss(w, i):
    x, y = trainExamples[i]
    return 2 * (w.dot(phi(x)) - y) * phi(x)

############################################################
# Optimization algorithm

def gradientDescent(F, gradientF, initialWeightVector):
    w = initialWeightVector()
    eta = 0.1
    for t in range(2):
        value = F(w)
        gradient = gradientF(w)
        print("Weights",w)
        #print("print(trainExamples)==>",trainExamples)
        print("eta",eta)
        print("gradient",gradient)
        w = w - eta * gradient
        print(f'epoch {t}: w = {w}, F(w) = {value}, gradientF = {gradient}')

def stochasticGradientDescent(f, gradientf, n, initialWeightVector):
    w = initialWeightVector()
    numUpdates = 0
    for t in range(2):
        for i in range(n):
            print("Weights",w)
            print("print(trainExamples)==>",trainExamples)
            value = f(w, i)
            gradient = gradientf(w, i)
            numUpdates += 1
            eta = 1.0 / math.sqrt(numUpdates)
            w = w - eta * gradient
            print("ttttttt",t)
        print(f'epoch {t}: w = {w}, F(w) = {value}, gradientF = {gradient}')

gradientDescent(trainLoss, gradientTrainLoss, initialWeightVector)
print("stochastic gradientd escent########################################################################")
stochasticGradientDescent(loss, gradientLoss, len(trainExamples), initialWeightVector)
```
